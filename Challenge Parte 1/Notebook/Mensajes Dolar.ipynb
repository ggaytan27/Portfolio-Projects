```python
urlCriptos = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page=1&sparkline=false'

responseCripto = requests.get(urlCriptos).json()

responseCripto

```

Dataframe

```python
def get_cripto(response, i):
    nombreCripto = response[i]['name']
    precioCripto = round(response[i]['current_price'], 2)
    rankingCripto = response[i]['market_cap_rank']

    return nombreCripto, precioCripto, rankingCripto
```

```python
dataCripto = []

for i in tqdm(range(len(responseCripto)), colour="green"):
    dataCripto.append(get_cripto(responseCripto, i))
```

```python
dataCripto[0]
```

```python
col = ['Moneda', 'Precio (USD)', 'Ranking']
dfCripto = pd.DataFrame(dataCripto, columns=col)
dfCripto
```

```python
dfCriptoFinal = dfCripto[dfCripto['Ranking']<=10]
```

```python
dfCriptoFinal

mensajeCripto = "Hola, resumen de las cripto: \n \n " + dfCriptoFinal.to_string(index=False)

mensajeCripto
```
